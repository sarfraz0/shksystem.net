```python
import feedparser
import pandas as pa
import numpy
import re
import requests
import json
import transmissionrpc as tr
from net.shksystem.common.send_mail import SendMail

regex = '^.HorribleSubs.\s([a-zA-Z0-9\s\'!-]+)\s-\s(\d{1,2}).*1080p.*$'
uri1  = 'https://getstrike.net/api/v2/torrents/search/?phrase=HorribleSubs%201080p&category=Anime'
uri2  = 'http://kat.cr/usearch/HorribleSubs%201080p/?rss=1'


class ReleaseFrame(object):
    """ This class implements getters for popular sites feeds """
    
    def _reord(self, d, regex, has_episodes=False, has_seasons=False):
        d = d.reindex(columns=['title', 'published', 'size', 'magnet_uri'])
        d.sort(['published', 'title', 'size'], ascending=False, inplace=True)

        def map_title(regex, group_num, default_ret, title):
            ret = default_ret
            ma = re.match(regex, title)
            if ma and re.compile(regex).groups >= group_num:
                ret = ma.group(group_num).strip()
            return ret

        if has_episodes and has_seasons:
            season_grp = 2
            episode_grp = 3
        elif has_episodes:
            season_grp = 110
            episode_grp = 2
        else:
            season_grp = 110
            episode_grp = 110

        d['name'] = d['title'] \
                .map(lambda x: map_title(regex, 1, numpy.nan, x))
        d.dropna(inplace=True)
        d['episode'] = d['title'] \
                .map(lambda x: int(map_title(regex, episode_grp, 0, x)))
        d['season'] = d['title'] \
                .map(lambda x: int(map_title(regex, season_grp, 0, x)))
        d.drop_duplicates(['name', 'season', 'episode'], inplace=True)

        return d
    
    def get_from_kickass(self, kickass_uri, regex, has_episodes=False, has_seasons=False):
        rss = feedparser.parse(kickass_uri)
        d = pa.DataFrame(rss['entries'])
        d.rename(columns={'torrent_magneturi': 'magnet_uri',
                          'torrent_contentlength': 'size'}, inplace=True)
        d = self._reord(d, regex, has_episodes, has_seasons)
        return d
    
    
    def get_from_strike(self, strike_uri, regex, has_episodes=False, has_seasons=False):
        r = requests.get(strike_uri)
        j = json.loads(r.text)
        d = pa.DataFrame(j['torrents'])
        d.rename(columns={'torrent_title': 'title', 'upload_date': 'published'},
                 inplace=True)
        d = self._reord(d, regex, has_episodes, has_seasons)
        return d
    
    def get_rule(self, d, name, ml_host, ml_user, ml_port=587, tr_host='localhost', tr_port=9091):
        f = d[d[name] == name]
        # Connecting to transmission and mail server
        ml = SendMail(ml_host, ml_port, ml_user)
        rc = tr.Client(tr_host, tr_port)
        for i, r in f.itterows():
            pass
        
 
        


feed1 = ReleaseFrame()
data2 = feed1.get_from_strike(uri1, regex, has_episodes=True)
data3 = feed1.get_from_kickass(uri2, regex, has_episodes=True)
data2
data3
#data2[data2['name'] == 'Tribe Cool Crew']
```
